# Задание 11

Для двух векторов:
С и b= {0.6; 1.2; -1.8, 1.05} найти меру косинусного сходства.


```python
b = np.array([0.6, 1.2, -1.8, 1.05], dtype=float).reshape(-1, 1)
cos_sim = (C.T @ b) / (np.linalg.norm(C) * np.linalg.norm(b))
cos_sim
```
